```python
# Import the necessary packages
import os
import numpy as np
from itertools import *
from more_itertools import *
from scipy import signal
from scipy.fft import fftshift
import matplotlib.pyplot as plt
%matplotlib inline

from sklearn.model_selection import train_test_split
import tensorflow.keras as keras
from keras import layers, models
from keras import utils
from keras.layers import Dense, Dropout, Activation, Flatten, Conv2D, MaxPooling2D,BatchNormalization,GRU,Bidirectional,Reshape
```

```python
fs = 16070;
windowIncrease = 0.1 #in sec
windowSize = 5 #in sec
filterOrder = 10
filterBand = [200,5000] #in Hertz


#audioCh1Data = np.ndarray([0,windowSize*fs])
#audioCh2Data = np.ndarray([0,windowSize*fs])
#Y = np.ndarray([0,2])
audioCh1Data = []
audioCh2Data = []
Y = []


for i in os.listdir("./../DataGarden"):           # iteration over all data files
    dist = int(i[:i.find("cm")])/500
    deg =  int(i[i.find("cm")+3:i.find("deg")])/180
    
    with open ("./../DataGarden/"+i,'r') as f:    # reading data file
        data = f.readlines ()
        data = [int(i) for i in data]
        #data = np.asarray(data)
    
    
        #l = int(np.size(data)/2)                  # split the two audio channels
        l = int(len(data)/2)
        audioCh1 = data[:l:]
        audioCh2 = data[l::]
        
        
        audioCh1 = audioCh1[2*fs:-2*fs:]          # remove 2 first and last seconds from data
        audioCh2 = audioCh2[2*fs:-2*fs:]
        
        
        sos = signal.butter(filterOrder, filterBand, 'bp', fs=fs, output='sos')
        audioCh1Filtered = signal.sosfilt(sos, audioCh1)
        audioCh2Filtered = signal.sosfilt(sos, audioCh2)
        
        #l = int(np.size(audioCh1Filtered))
        l = int(len(audioCh1Filtered))
        
        #tempData = np.asarray(list(windowed(list(audioCh1Filtered), n=int(windowSize*fs), step=int(windowIncrease*fs))))
        #audioCh1Data = np.append(audioCh1Data,tempData[0:-1,:],axis=0)
        
        #tempData = np.asarray(list(windowed(list(audioCh2Filtered), n=int(windowSize*fs), step=int(windowIncrease*fs))))
        #audioCh2Data = np.append(audioCh2Data,tempData[0:-1,:],axis=0)
        
        #Y = np.append(Y,np.asarray(list(repeat([dist,deg],np.size(tempData[0:-1,0])))),axis=0)
        
        tempData = list(windowed(audioCh1Filtered, n=int(windowSize*fs), step=int(windowIncrease*fs)))
        tempData.pop()
        audioCh1Data.extend(tempData)
        
        tempData = list(windowed(audioCh2Filtered, n=int(windowSize*fs), step=int(windowIncrease*fs)))
        tempData.pop()
        audioCh2Data.extend(tempData)
        
        dataSize = (np.shape(tempData))[0]
        Y.extend(list(repeat([dist,deg],dataSize))) 
        
        
        print (np.shape(audioCh1Data),dist,deg)
    f.close()
```

```python
audioCh1Data = np.asarray(audioCh1Data)
audioCh2Data = np.asarray(audioCh2Data)
Y = np.asarray(Y)
np.save ('Ch1.npy',audioCh1Data)
np.save ('Ch2.npy',audioCh2Data)
np.save ('Y.npy',Y)
```

```python
audioCh1Data = np.float16(np.load ('Ch1.npy')/32768)
audioCh2Data = np.float16(np.load ('Ch2.npy')/32768)
Y = np.load ('Y.npy')
fs = 16070;
```

```python
print (np.shape(audioCh1Data))
print (np.shape(audioCh2Data))
print (np.shape(Y))
print (audioCh1Data.dtype)
```

```python
f, t, MagCh1 = signal.spectrogram(audioCh1Data, fs,window='hamming',scaling='spectrum',mode='magnitude')
f, t, MagCh2 = signal.spectrogram(audioCh2Data, fs,window='hamming',scaling='spectrum',mode='magnitude')
f, t, PhaseCh1 = signal.spectrogram(audioCh1Data, fs,window='hamming',scaling='spectrum',mode='angle')
f, t, PhaseCh2 = signal.spectrogram(audioCh2Data, fs,window='hamming',scaling='spectrum',mode='angle')

MagCh1 = np.float16(MagCh1)
MagCh2 = np.float16(MagCh2)
PhaseCh1 = np.float16(PhaseCh1)
PhaseCh2 = np.float16(PhaseCh2)

# print (np.average(MagCh1[0:87])   ,np.average(MagCh2[0:87]))
# print (np.average(MagCh1[87:171]) ,np.average(MagCh2[87:171]))
# print (np.average(MagCh1[171:281]),np.average(MagCh2[171:281]))
# print (np.average(MagCh1[281:361]),np.average(MagCh2[281:361]))
# print (np.average(MagCh1[361:442]),np.average(MagCh2[361:442]))
# print (np.average(MagCh1[442:530]),np.average(MagCh2[442:530]))
# print (np.average(MagCh1[530:619]),np.average(MagCh2[530:619]))
# print (np.average(MagCh1[619:705]),np.average(MagCh2[619:705]))
# print (np.average(MagCh1[705:794]),np.average(MagCh2[705:794]))
# print (np.average(MagCh1[794:896]),np.average(MagCh2[794:896]))
# print (np.average(MagCh1[896:992]),np.average(MagCh2[896:992]))
# print (np.average(MagCh1[992:1079]),np.average(MagCh2[992:1079]))
# print (np.average(MagCh1[1079:1160]),np.average(MagCh2[1079:1160]))
# print (np.average(MagCh1[1160:1243]),np.average(MagCh2[1160:1243]))
# print (np.average(MagCh1[1243:1351]),np.average(MagCh2[1243:1351]))

index = 0
plt.figure(figsize=(12, 8))
plt.subplot (221)
plt.pcolormesh(t, f, MagCh1[index])
plt.ylabel('Frequency [Hz]')
plt.xlabel('Time [sec]')
plt.subplot (222)
plt.pcolormesh(t, f, PhaseCh1[index])
plt.ylabel('Frequency [Hz]')
plt.xlabel('Time [sec]')
plt.subplot (223)
plt.pcolormesh(t, f, MagCh2[index])
plt.ylabel('Frequency [Hz]')
plt.xlabel('Time [sec]')
plt.subplot (224)
plt.pcolormesh(t, f, PhaseCh2[index])
plt.ylabel('Frequency [Hz]')
plt.xlabel('Time [sec]')
plt.show()
```

```python
print (np.shape(MagCh1))
print (np.shape(MagCh2))
print (MagCh1.dtype)
print (np.shape(PhaseCh1))
print (np.shape(PhaseCh2))
print (PhaseCh1.dtype)
dataSize = list(np.shape(MagCh1))
dataSize.append(4)
X = np.ndarray(dataSize, dtype=np.float16)
print (np.shape(X))
X[:,:,:,0] = MagCh1
X[:,:,:,1] = MagCh2
X[:,:,:,2] = PhaseCh1
X[:,:,:,3] = PhaseCh2
```

```python
X_train, X_validate, Y_train, Y_validate = train_test_split(X, Y, test_size = 0.2, random_state = 2)
print (np.shape(X_train))
print (np.shape(X_validate))
print (np.shape(Y_train))
print (np.shape(Y_validate))
```

```python
# Define batchsize and number of epochs
batchSize = 20
nEpochs = 20


# Declare the sequential model and design your multi-layer perceptron
model = models.Sequential()

model.add(Conv2D(8, (3, 3), padding='same',
                 input_shape=X_train.shape[1:]))
model.add(Activation('relu'))
model.add(BatchNormalization())
model.add(MaxPooling2D(pool_size=(2, 2)))

model.add(Conv2D(16, (3, 3), padding='same'))
model.add(Activation('relu'))
model.add(BatchNormalization())
model.add(MaxPooling2D(pool_size=(2, 2)))

model.add(Conv2D(32, (3, 3), padding='same'))
model.add(Activation('relu'))
model.add(BatchNormalization())
model.add(MaxPooling2D(pool_size=(2, 2)))


# model.add(Conv2D(64, (3, 3), padding='same'))
# model.add(Activation('relu'))
# model.add(BatchNormalization())
# model.add(MaxPooling2D(pool_size=(2, 2)))

# model.add(Conv2D(128, (3, 3), padding='same'))
# model.add(Activation('relu'))
# model.add(BatchNormalization())
# model.add(MaxPooling2D(pool_size=(2, 2)))

model.add(Reshape((16,1408)))
model.add(Bidirectional(GRU(32)))

#model.add(Bidirectional(GRU(64)))

#model.add(Flatten())
model.add(Dense(256))
model.add(Activation('linear'))
model.add(Dense(64))
model.add(Activation('linear'))
model.add(Dense(1))
model.add(Activation('sigmoid'))

# Compile your model
model.compile(loss='mean_absolute_error',
              optimizer='sgd',
              metrics=['mae','mse'])

X_train = X_train.astype('float16')
X_validate = X_validate.astype('float16')
Y_train = Y_train.astype('float16')
Y_validate = Y_validate.astype('float16')

# Train your model
model.fit(X_train, Y_train[:,0], epochs = nEpochs , batch_size = batchSize)
```

```python
model.build()
model.summary ()
```

```python
model.evaluate (X_validate,Y_validate[:,0])
```

```python
Y_pred = model.predict (X_validate)
index = 0
plt.plot (Y_pred[:,0],'bs',Y_validate[:,0],'g^')
plt.show()
```
